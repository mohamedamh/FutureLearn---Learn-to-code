```python
import warnings
warnings.simplefilter('ignore', FutureWarning)

from pandas import *
moscow = read_csv('Moscow_2014.csv', skipinitialspace=True)
```

## Cleaning the data

```python
moscow = moscow.rename(columns={'WindDirDegrees<br />' : 'WindDirDegrees'})
```

```python
moscow['WindDirDegrees'] = moscow['WindDirDegrees'].str.rstrip('<br />')
```

```python
moscow['WindDirDegrees'] = moscow['WindDirDegrees'].astype('float64')   
```

```python
moscow['GMT+3'] = to_datetime(moscow['GMT+3'])
```

```python
moscow.index = moscow['GMT+3']
```

## Finding a summer break

According to meteorologists,June, July and August are summer months in Moscow. It is hot or warm. The days are long and the nights are short.
December, January and February are winter months. The weather is cold. Usually it snows. The days are short and the nights are long.

```python
summer = moscow.loc[datetime(2014,6,1) : datetime(2014,8,31)]
```

```python
summer[summer['Mean TemperatureC'] >= 25]
```

There are only two days the Temperature higher than 25°C. 

```python
%matplotlib inline
```

```python
summer['Mean TemperatureC'].plot(grid=True, figsize=(10,5))
```

In July the heat begins to rise to reach the peak at the end of July and the beginning of August.

```python
summer[['Mean TemperatureC', 'Precipitationmm']].plot(grid=True, figsize=(10,5))
```

No Precipitation no rain,The best period for holiday in summer is the end of July and the beginning of August.

```python
summer[['Mean TemperatureC', 'Precipitationmm']][datetime(2014,7,24) : datetime(2014,8,7)].plot(grid=True, figsize=(10,5))
```

## Conclusion

During the long summer days, typical high temperatures in the warm months of June, July and August are around 23 °C, and we observe the absence of precipitation.
